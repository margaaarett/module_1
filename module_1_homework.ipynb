# 14. Какая команда выиграла наибольшое количество матчей в сезоне? (Будьте внимательны с типом матча в сезоне - season).

```python
season_r = data[data['season'] == 'regular season'] #На сколько я понимаю надо считать именно regular season
home_wins = season_r[season_r['home_team_win'] == 1]['home_team'].value_counts() # Победа домашних команд
away_wins = season_r[season_r['home_team_loss'] == 1]['away_team'].value_counts() #Прогрыш домашней команды означает выйгрыш гостевой команлы
total_wins=home_wins.add(away_wins, fill_value=0).astype(int)
top_team = total_wins.idxmax()
print(f"Команда, выйгравшая наибольшее кол-во матчей: {top_team}")
```

# 18. Зависит ли выигрыш от количества посетителей матча? (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.cov.html)

```python
#Взяла home_team_win, т к для того, чтобы посчитать коварацию нужны числовые параметры 
data_at=data[['home_team_win','attendance']].dropna()
x=data_at['home_team_win'].to_numpy()
y=data_at['attendance'].to_numpy()
covariance = np.cov(x,y,ddof=1)[0,1]
print(f"Ковариация: {covariance}")
```

```python
#Интерпритация: значение ковариации не дает никакого толкования, поэтому посчитаем разницу средних из ковариации. если р=0.5 И diff=903, то в среднем на победные матчи приходило на 900 зрителей больше
p=data['home_team_win'].mean()
diff=covariance/(p*(1-p))
print(p,diff)
```
