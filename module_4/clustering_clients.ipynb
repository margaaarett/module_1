##### Демографические данные

    OnlinePurchases: Количество покупок, совершенных через веб-сайт компании.

```python
# Выбросы - метод IQR (межквартильный размах)
import numpy as np

def cap_outliers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return col[(col < lower) | (col > upper)]

# Посмотрим количество выбросов для ключевых числовых признаков
cols_to_check = ["AnnualIncome", "SpentOnWines", "SpentOnFruits",
                 "SpentOnMeat", "SpentOnFish", "SpentOnSweets", "SpentOnGold"]

for col in cols_to_check:
    if col in df.columns:
        outliers = cap_outliers(df[col])
        print(col, "- выбросов:", outliers.shape[0])
```

```python
# Простой вариант обработки выбросов — "обрезка" (winsorization) по границам IQR

def clip_outliers_iqr(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return col.clip(lower=lower, upper=upper)

# Применим обрезку для основных числовых признаков
for col in cols_to_check:
    if col in df.columns:
        df[col] = clip_outliers_iqr(df[col])
```
